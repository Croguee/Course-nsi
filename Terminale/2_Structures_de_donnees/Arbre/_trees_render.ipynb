```python
from graphviz import Digraph
from IPython.display import display

tree = Digraph(format='svg')
tree.node('A')
tree.node('B')
tree.node('C')
tree.node('D')
tree.node('E')
tree.node('F')
tree.node('G')
tree.edges(['AB','AC','AD','BE','BF','CG'])
tree.render('img/tree.gv')
```

```python
tree.node('E', color='red', fontcolor='red')
tree.node('B', color='red', fontcolor='red')
tree.node('F', color='red', fontcolor='red')
tree.render('img/sub-tree.gv')
```

```python
bstree=Digraph(format='svg', comment='Arbre binaire de recherche')
bstree.attr(ordering="out")
for value in range(1,10):
    bstree.node(str(value),str(value), shape='circle')
bstree.node('A','')
bstree.node('B','')
bstree.edges(['52','21','24','43','4A','58','86','6B','67','89'])
bstree.node('A', shape='point', fillcolor="white")
bstree.node('B', shape='point', fillcolor="white")
display(bstree)
bstree.render('img/bstree.gv')
```

```python
bstree.node('5', fillcolor='red', style='filled')
bstree.node('4', fillcolor='green', style='filled')
bstree.node('6', fillcolor='green', style='filled')
display(bstree)
bstree.render('img/bstree2.gv')
```

```python
bstree_l = bstree.copy()
bstree_l.node('5',label = '4', fillcolor='green')
bstree_l.node('6', fillcolor='white')
bstree_l.node('4', label='3', fillcolor='white')
bstree_l.node('3', label='', shape='point', fillcolor='green')

display(bstree_l)
bstree_l.render('img/bstree_l.gv')
```

```python

bstree_r = bstree.copy()
bstree_r.node('5',label = '6', fillcolor='green')
bstree_r.node('4',fillcolor='white')
bstree_r.node('6',label = '7', fillcolor='white')
bstree_r.node('7',label = '', fillcolor='green', shape='point')

display(bstree_r)
bstree_r.render('img/bstree_r.gv')
```
