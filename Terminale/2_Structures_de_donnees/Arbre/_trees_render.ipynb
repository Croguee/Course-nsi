```python
from graphviz import Digraph
from IPython.display import display

tree = Digraph(format='svg')
tree.node('A')
tree.node('B')
tree.node('C')
tree.node('D')
tree.node('E')
tree.node('F')
tree.node('G')
tree.edges(['AB','AC','AD','BE','BF','CG'])
tree.render('img/tree.gv')
```

```python
tree.node('E', color='red', fontcolor='red')
tree.node('B', color='red', fontcolor='red')
tree.node('F', color='red', fontcolor='red')
tree.render('img/sub-tree.gv')
```

```python
heigh_depth=Digraph(format='svg', comment='Hauteur et profondeur')
heigh_depth.node('A','p=0, h=3')
heigh_depth.node('B', 'p=1, h=2')
heigh_depth.node('C', 'p=1, h=2')
heigh_depth.node('D', 'p=2, h=1')
heigh_depth.node('E', 'p=2, h=1')
heigh_depth.node('F', 'p=2, h=1')
heigh_depth.node('G', 'p=3, h=0')
heigh_depth.node('H', 'p=3, h=0')
heigh_depth.edges(['AB','AC','BD','CE','CF','EG', 'EH'])
display(heigh_depth)
heigh_depth.render('img/heigh_depth.gv')
```
